## Exercise 1

```r
# Load the CSV file
EPI_data <- read.csv("epi2024results06022024.csv")

# Exploring the 'EPI.new' variable
# Summary statistics
summary(EPI_data$EPI.new)

# Five-number summary
fivenum(EPI_data$EPI.new, na.rm = TRUE)

# Stem and Leaf Plot
stem(EPI_data$EPI.new)

# Histogram
hist(EPI_data$EPI.new, main = "Histogram of EPI.new", xlab = "EPI Score", col = "lightblue")

# Histogram with density line
hist(EPI_data$EPI.new, seq(20, 80, 1), prob = TRUE, main = "Histogram with Density", xlab = "EPI Score", col = "lightgreen")
lines(density(EPI_data$EPI.new, na.rm = TRUE, bw = 1), col = "red")
rug(EPI_data$EPI.new)

# Cumulative density function
plot(ecdf(EPI_data$EPI.new), do.points = FALSE, verticals = TRUE, main = "Cumulative Density Function", xlab = "EPI Score")

# Quantile-Quantile Plot
qqnorm(EPI_data$EPI.new, main = "Q-Q Plot")
qqline(EPI_data$EPI.new)
```

## Exercise 2

```r
# Load the CSV file
EPI_data <- read.csv("epi2024results06022024.csv")

# Handle missing values
EPI_data$EPI.new <- na.omit(EPI_data$EPI.new)

# Cumulative Density Function (CDF)
plot(ecdf(EPI_data$EPI.new), do.points = FALSE, verticals = TRUE,
     main = "Cumulative Density Function", xlab = "EPI Score")

# Quantile-Quantile Plot
qqnorm(EPI_data$EPI.new, main = "Q-Q Plot for Normal Distribution")
qqline(EPI_data$EPI.new, col = "blue")

# Q-Q Plot against the Generating Normal Distribution
qqplot(rnorm(250), EPI_data$EPI.new, 
       xlab = "Theoretical Quantiles (Normal Dist.)", 
       ylab = "Sample Quantiles", 
       main = "Q-Q Plot for Normal Distribution")
qqline(EPI_data$EPI.new, col = "red")

# Q-Q Plot against the t-Distribution
qqplot(rt(250, df = 5), EPI_data$EPI.new, 
       xlab = "Theoretical Quantiles (t-Distribution)", 
       ylab = "Sample Quantiles", 
       main = "Q-Q Plot for t Distribution")
qqline(EPI_data$EPI.new, col = "darkgreen")
```

## Exercise 2a

```r
# Load the CSV file
EPI_data <- read.csv("epi2024results06022024.csv")

# Handle missing values for both variables (APO.new and WRS.new)
EPI_data$APO.new <- na.omit(EPI_data$APO.new)
EPI_data$WRS.new <- na.omit(EPI_data$WRS.new)

# Cumulative Density Function (CDF) for APO.new
plot(ecdf(EPI_data$APO.new), do.points = FALSE, verticals = TRUE,
     main = "Cumulative Density Function for APO.new", xlab = "APO.new Score")

# Q-Q Plot for APO.new
qqnorm(EPI_data$APO.new, main = "Q-Q Plot for APO.new (Normal Distribution)")
qqline(EPI_data$APO.new, col = "blue")

# Q-Q Plot for APO.new against a Normal Distribution
qqplot(rnorm(250), EPI_data$APO.new,
       xlab = "Theoretical Quantiles (Normal Dist.)",
       ylab = "Sample Quantiles",
       main = "Q-Q Plot for APO.new (Normal Distribution)")
qqline(EPI_data$APO.new, col = "red")

# Q-Q Plot for APO.new against t-Distribution
qqplot(rt(250, df = 5), EPI_data$APO.new,
       xlab = "Theoretical Quantiles (t-Distribution)",
       ylab = "Sample Quantiles",
       main = "Q-Q Plot for APO.new (t-Distribution)")
qqline(EPI_data$APO.new, col = "darkgreen")

# Repeat the same for WRS.new
# CDF for WRS.new
plot(ecdf(EPI_data$WRS.new), do.points = FALSE, verticals = TRUE,
     main = "Cumulative Density Function for WRS.new", xlab = "WRS.new Score")

# Q-Q Plot for WRS.new
qqnorm(EPI_data$WRS.new, main = "Q-Q Plot for WRS.new (Normal Distribution)")
qqline(EPI_data$WRS.new, col = "blue")

# Q-Q Plot for WRS.new against a Normal Distribution
qqplot(rnorm(250), EPI_data$WRS.new,
       xlab = "Theoretical Quantiles (Normal Dist.)",
       ylab = "Sample Quantiles",
       main = "Q-Q Plot for WRS.new (Normal Distribution)")
qqline(EPI_data$WRS.new, col = "red")

# Q-Q Plot for WRS.new against t-Distribution
qqplot(rt(250, df = 5), EPI_data$WRS.new,
       xlab = "Theoretical Quantiles (t-Distribution)",
       ylab = "Sample Quantiles",
       main = "Q-Q Plot for WRS.new (t-Distribution)")
qqline(EPI_data$WRS.new, col = "darkgreen")
```
